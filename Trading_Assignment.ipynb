```python
# Convert trader timestamp to datetime
trades["Timestamp IST"] = pd.to_datetime(
    trades["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)

# Create a Date column for merging
trades["Date"] = trades["Timestamp IST"].dt.date

# Check the data
trades.head()
```

```python
# Convert date column to datetime
sentiment["date"] = pd.to_datetime(sentiment["date"])

# Create a matching Date column
sentiment["Date"] = sentiment["date"].dt.date

sentiment.head()
```

```python
merged = pd.merge(
    trades,
    sentiment,
    on="Date",
    how="left"
)
```

```python
merged.head()
```

```python
merged[merged["classification"].isnull()][["Date"]].drop_duplicates()
```

```python
merged = merged.dropna(subset=["classification"])
```

```python
merged["classification"].isnull().sum()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=merged,
    x="classification",
    order=merged["classification"].value_counts().index
)

plt.title("Number of Trades by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")

plt.xticks(rotation=20)

plt.show()
```

This chart shows how trading activity varies across different market sentiment categories. A higher number of trades during a particular sentiment indicates increased trader participation under those market conditions.

```python
plt.figure(figsize=(8,5))

avg_pnl = merged.groupby("classification")["Closed PnL"].mean().sort_values()

sns.barplot(
    x=avg_pnl.index,
    y=avg_pnl.values
)

plt.title("Average Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Closed PnL")

plt.xticks(rotation=20)

plt.show()
```

This visualization compares the average profit or loss generated under different market sentiment conditions. It helps identify whether traders tend to perform better during Fear, Greed, or Extreme Fear periods.

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=merged,
    x="classification",
    y="Closed PnL"
)

plt.title("Profit Distribution by Market Sentiment")

plt.xticks(rotation=20)

plt.show()
```

```python
merged["Win"] = merged["Closed PnL"] > 0
```

```python
win_rate = (
    merged.groupby("classification")["Win"]
    .mean()
    .sort_values(ascending=False) * 100
)

plt.figure(figsize=(8,5))

sns.barplot(
    x=win_rate.index,
    y=win_rate.values
)

plt.title("Win Rate by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate (%)")

plt.xticks(rotation=20)

plt.tight_layout()
plt.savefig("chart4_win_rate.png", dpi=300)
plt.show()
```

This chart compares the percentage of profitable trades across different market sentiment conditions. A higher win rate suggests that traders were more successful under that sentiment.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=merged,
    x="Direction",
    hue="classification"
)

plt.title("Trade Direction by Market Sentiment")
plt.xlabel("Trade Direction")
plt.ylabel("Number of Trades")

plt.tight_layout()
plt.savefig("chart5_direction.png", dpi=300)
plt.show()
```

This chart shows whether traders preferred long (Buy) or short (Sell) positions during different market sentiment phases.

```python
volume = (
    merged.groupby("classification")["Size USD"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sns.barplot(
    x=volume.index,
    y=volume.values
)

plt.title("Total Trading Volume by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Total Size (USD)")

plt.xticks(rotation=20)

plt.tight_layout()
plt.savefig("chart6_volume.png", dpi=300)
plt.show()
```

Higher trading volume during a particular market sentiment indicates stronger market participation and potentially increased volatility.

```python
top_coins = (
    merged.groupby("Coin")["Closed PnL"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_coins.values,
    y=top_coins.index
)

plt.title("Top 10 Coins by Total Closed PnL")
plt.xlabel("Total Closed PnL")
plt.ylabel("Coin")

plt.tight_layout()
plt.savefig("chart7_top_coins.png", dpi=300)
plt.show()
```

This chart highlights which trading pairs generated the highest cumulative profits across all trades.

```python
top_accounts = (
    merged.groupby("Account")["Closed PnL"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_accounts.values,
    y=top_accounts.index
)

plt.title("Top 10 Trader Accounts by Closed PnL")
plt.xlabel("Total Closed PnL")
plt.ylabel("Account")

plt.tight_layout()
plt.savefig("chart8_top_accounts.png", dpi=300)
plt.show()
```

A small number of trader accounts may contribute disproportionately to total profits, indicating differences in strategy or trading expertise.

```python
# Make sure it's datetime
merged["Timestamp IST"] = pd.to_datetime(
    merged["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)

# Create Hour column
merged["Hour"] = merged["Timestamp IST"].dt.hour

# Create Weekday column (we'll use it later)
merged["Weekday"] = merged["Timestamp IST"].dt.day_name()
```

```python
merged[["Timestamp IST", "Hour", "Weekday"]].head()
```

```python
hourly = merged.groupby("Hour").size()

plt.figure(figsize=(10,5))

hourly.plot(marker="o")

plt.title("Trading Activity by Hour (IST)")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trades")

plt.grid(True)
plt.tight_layout()
plt.savefig("chart9_hourly.png", dpi=300)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = [
    "Execution Price",
    "Size Tokens",
    "Size USD",
    "Closed PnL",
    "Fee",
    "Start Position"
]

corr = merged[numeric_cols].corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.tight_layout()
plt.savefig("chart10_heatmap.png", dpi=300)

plt.show()
```

The correlation heatmap identifies relationships between numerical variables. Strong positive or negative correlations can reveal factors associated with larger profits, trading volume, or fees.

```python
daily = merged.groupby("Date")["Closed PnL"].sum()

plt.figure(figsize=(15,5))

daily.plot()

plt.title("Daily Closed PnL")
plt.xlabel("Date")
plt.ylabel("Total Closed PnL")

plt.grid(True)

plt.tight_layout()
plt.savefig("chart11_daily_pnl.png", dpi=300)

plt.show()
```

Daily profitability fluctuates over time, reflecting changing market conditions and trading performance.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=merged,
    x="Fee",
    y="Closed PnL"
)

plt.title("Fee vs Closed PnL")

plt.tight_layout()
plt.savefig("chart12_fee_vs_pnl.png", dpi=300)

plt.show()
```

This chart examines whether higher trading fees are associated with larger profits, which may indicate that larger or more active trades generate both higher costs and higher returns.

```python
plt.figure(figsize=(10,5))

sns.histplot(
    merged["Closed PnL"],
    bins=50,
    kde=True
)

plt.title("Distribution of Closed PnL")

plt.tight_layout()
plt.savefig("chart13_distribution.png", dpi=300)

plt.show()
```

The distribution of Closed PnL highlights whether most trades result in small gains/losses or whether a small number of large trades dominate overall performance.

##  Explore the relationship between trader performance and market sentiment.

Number of Trades by Market Sentiment
Observation

The highest number of trades occurred during Fear, with more than 60,000 trades. Greed was the second most active market condition, while Extreme Fear recorded the lowest trading activity. This indicates that traders remained highly active during fearful market conditions instead of avoiding the market.

Business Insight

The data suggests that traders viewed Fear periods as potential buying opportunities. Trading platforms and strategy developers should pay closer attention to Fear markets, as they attract significant trading activity.

Profit Distribution by Market Sentiment
Observation

The distribution shows that most trades generated relatively small profits or losses, while a few trades resulted in exceptionally large gains and losses. Extreme Greed and Fear contain several high-profit outliers, whereas Extreme Fear includes some of the largest losses.

Business Insight

A small number of trades contribute significantly to overall profitability. Effective risk management is important because a few large losing trades can offset many profitable ones.

Win Rate by Market Sentiment

Business Insight

Traders were more likely to close profitable trades during Extreme Greed, while Extreme Fear appeared to be the most challenging market condition.

Trade Direction by Market Sentiment
Observation

Buy and Long positions dominated trading activity across most market conditions. During Fear, traders opened significantly more Long positions, whereas Greed showed balanced participation between Long and Short trades.

Business Insight

Most traders preferred taking long positions rather than short positions, suggesting a general expectation of future price appreciation even during uncertain market conditions.

Total Trading Volume by Market Sentiment
Observation

Fear recorded the highest total trading volume, followed by Greed. Extreme Fear had the lowest trading volume, indicating reduced market participation during periods of maximum uncertainty.

Business Insight

Fear markets attracted the largest amount of capital. This indicates that experienced traders actively traded during market declines instead of waiting on the sidelines.

Top 10 Coins by Total Closed PnL
Observation

The @107 token generated the highest cumulative profit, followed by HYPE, SOL, ETH, and BTC. These assets contributed the majority of the total realized profits in the dataset.

Business Insight

Trading profits were concentrated in a few high-performing assets. Focusing on consistently profitable coins may improve portfolio performance.

Top 10 Trader Accounts
Observation

Business Insight

The results indicate that profitability is concentrated among a limited number of highly successful traders, suggesting differences in trading strategy, experience, or risk management.

Trading Activity by Hour
Observation

Trading activity increased significantly during the evening hours, peaking around 7 PM to 8 PM IST. Activity was comparatively lower during the afternoon.

Business Insight

The evening session appears to be the busiest trading period. Traders may be responding to global market activity or increased liquidity during these hours.

Correlation Heatmap
Observation

The strongest relationship is between Size USD and Fee (0.75), indicating that larger trades naturally incur higher transaction costs. Closed PnL has only weak correlations with the other variables, suggesting that profitability depends on multiple factors rather than any single feature.

Business Insight

Trade size strongly influences fees, but profit cannot be explained solely by trade size, execution price, or starting position. Successful trading likely depends on strategy, market timing, and risk management.

Daily Closed PnL
Observation

Daily profits remained relatively stable during the early period but became much more volatile from late 2024 onward. Several days recorded exceptionally high profits as well as substantial losses.

Business Insight

Profitability fluctuated considerably over time, reflecting changing market conditions. Traders should adapt their risk exposure during periods of increased volatility.

Fee vs Closed PnL
Observation

Most trades had relatively low transaction fees, and there is no clear linear relationship between fees and profitability. While a few high-fee trades generated substantial profits, others resulted in losses.

Business Insight

Paying higher transaction fees does not guarantee better returns. Trading success depends more on execution quality and market conditions than on trade size alone.

Observation

Business Insight

The overall profit distribution is highly skewed, meaning a small number of exceptional trades contribute disproportionately to total profitability. Consistent risk management is therefore essential.
